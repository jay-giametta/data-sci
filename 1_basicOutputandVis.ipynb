# Basic Output and Visualization


Import necessary libraries

```
import pandas as pd
import matplotlib.pyplot as plt
```

Read the Iris dataset into a dataframe

```
iris = pd.read_csv("https://datahub.io/machine-learning/iris/r/iris.csv")
```

Print the first few observations in the dataset

```
iris.head()
```

Set default plot size


```
plt.rcParams['figure.figsize'] = [12, 8]
```

Plot a scatter matrix from the dataframe

```
fig1 = pd.plotting.scatter_matrix(iris, alpha=0.2)
```

Plot sepal width/length of each class

```
fig2, ax1 = plt.subplots()
for cls in iris["class"].unique():
    plt.scatter(x=iris.loc[iris["class"] == cls]['sepalwidth'],
                y=iris.loc[iris["class"] == cls]['sepallength'],
                label=cls)
ax1.legend()
ax1.grid()
```
