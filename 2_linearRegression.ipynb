# Linear Regression

Import necessary libraries

```
import sklearn.linear_model as lm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read the Iris dataset into a dataframe

```
iris = pd.read_csv("https://datahub.io/machine-learning/iris/r/iris.csv")
```

Choose feature values

```
features = iris[['petallength']]
```

Choose target values

```
target = iris.petalwidth
```

Instantiate a regression model

```
model = lm.LinearRegression()
```

Fit the model using the entire dataset.  

Note: Will cover using separate training/test sets and validation in a later notebook



```
model.fit(features, target)
```

Save the linear equation from the model

```
line = 'y = ' + str(round(model.coef_[0], 2)) + "x + " + str(round(model.intercept_, 2))
print(line)
```

Make predictions with the fitted model

```
predict = model.predict(features)
```

Calculate model accuracy using root mean square error (RMSE)

```
se = (target - predict)**2
mse = se.mean()
rmse = round(np.sqrt(mse), 4)
print(rmse)
```

Plot predictions vs truth data

```
# set default plot size
plt.rcParams['figure.figsize'] = [12, 8]

# plot predictions
plt.plot(predict, label='prediction')

# plot truth data
plt.plot(target, label='truth')

# pretty graph settings
plt.suptitle("Linear Regression: " + line)
plt.title("RMSE: " + str(rmse))
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
plt.grid()
```
